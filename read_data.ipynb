## Read ppg data

```python
import os
import torch

ppg_sig = []
for subject in range(1, 220):
    subjectid = quality_dict[subject][0]
    segment = seg_choose[subject-1]
    path = os.path.join(ppg_root, '{}_{}.txt'.format(subjectid, segment))
    if os.path.exists(path):
      with open(path) as f:
        lines = f.readlines()[0].split('\t')[:-1]
        if len(lines) != 2100:
          print(subject, subjectid, segment, len(lines))
          continue
        lines = [float(x) for x in lines]
        ppg_sig.append(lines)

print(len(ppg_sig))
```

```python
import torch

# convert list to tensor
ppg_sig = torch.Tensor(ppg_sig)
ppg_sig.size()
```

## Read BP label
Label:
- 'Stage 1 hypertension': 1
- 'Stage 2 hypertension': 2
- 'Prehypertension': 3
- 'Normal': 0

```python
import pandas as pd

bp_label = []
df = pd.read_csv(label_root, skiprows=1)
print(set(df["Hypertension"]))
for subject in range(219):
  if subject == 180:
    continue
  if df['Hypertension'][subject] == 'Stage 1 hypertension':
    bp_label.append(1)
  elif df['Hypertension'][subject] == 'Stage 2 hypertension':
    bp_label.append(2)
  elif df['Hypertension'][subject] == 'Prehypertension':
    bp_label.append(3)
  elif df['Hypertension'][subject] == 'Normal':
    bp_label.append(0)
```

## Turn BP label into tensor

```python
# turn bp label into tensor

bp_label = torch.Tensor(bp_label)
bp_label.size()
```
